```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
# pip install folium
import folium
```

```python
# folium의 Map() 메소드의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 속성으로 지정한 배율을 사용하는
# 지도를 만든다.
# folium.Map(location=[위도, 경도], zoom_start=배율[, tiles='지도 스타일'])
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)
# folium 지도 객체에서 save() 메소드로 지도 차트를 html 파일로 저장할 수 있다.
folium_map.save('./output/map.html')
folium_map
```

Map() 메소드의 tiles 속성으로 지도 스타일을 지정할 수 있다.  
지도 스타일: Open street Map(기본값), Stamen Terrain, Stamen Toner

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15, tiles='Open street Map')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15, tiles='Stamen Terrain')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15, tiles='Stamen Toner')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# folium의 Marker() 메소드의 location 속성으로 지정한 위치에 마커를 만든다. => add_to() 메소드로 지도에 추가한다.
# folium의 Marker() 메소드의 popup 속성으로 마커를 클릭했을 때 표시할 문자열을 지정할 수 있다.
folium.Marker(location=[37.579617, 126.977041], popup='경복궁').add_to(folium_map)
folium.Marker(location=[37.5658049, 126.9751461], popup='덕수궁').add_to(folium_map)
folium.Marker(location=[37.5794309, 126.9910426], popup='창덕궁').add_to(folium_map)
folium.Marker(location=[37.5787708, 126.9948698], popup='창경궁').add_to(folium_map)
folium.Marker(location=[37.57327406, 126.99516214], popup='종묘').add_to(folium_map)

# 마커를 클릭했을 때 팝업되는 말풍선의 크기는 folium의 Popup() 메소드의 max_width 속성으로 변경할 수 있다.
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.5697869, 126.9823161], popup=popup).add_to(folium_map)

folium_map.save('./output/map2.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# Marker() 메소드의 icon 속성에 Icon() 메소드로 마커의 색상과 아이콘 모양을 지정할 수 있다.
# 마커에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 아이콘을
# 표시할 수 있다.
# 색상 이름 리스트
# 'blue', 'cadetblue', 'pink', 'red', 'lightgreen', 'purple', 'lightred', 'beige', 'orange', 'black', 'darkblue'
# 'white', 'gray', 'darkred', 'darkpurple', 'green', 'darkgreen', 'lightgray', 'lightblue'

folium.Marker(location=[37.579617, 126.977041], popup='경복궁',
             icon=folium.Icon(color='red', icon='arrow-down')).add_to(folium_map)
folium.Marker(location=[37.5658049, 126.9751461], popup='덕수궁',
             icon=folium.Icon(color='green', icon='ok')).add_to(folium_map)
folium.Marker(location=[37.5794309, 126.9910426], popup='창덕궁',
             icon=folium.Icon(color='purple', icon='star')).add_to(folium_map)
folium.Marker(location=[37.5787708, 126.9948698], popup='창경궁',
             icon=folium.Icon(color='orange', icon='heart')).add_to(folium_map)
folium.Marker(location=[37.57327406, 126.99516214], popup='종묘',
             icon=folium.Icon(color='pink', icon='cutlery')).add_to(folium_map)

folium_map.save('./output/map3.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# folium의 CircleMarker() 메소드로 원형 마커를 만들 수 있다.
# radius 속성으로 마커로 표시할 원의 반지름을 지정한다.
# color 속성으로 원의 색상, fill_color 속성으로 원 내부의 색상을 지정한다.

folium.CircleMarker(location=[37.579617, 126.977041], popup='경복궁',
                   radius=100, color='red', fill_color='blue').add_to(folium_map)
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.CircleMarker(location=[37.5697869, 126.9823161], popup=popup,
                   radius=50, color='#48cfde', fill_color='#ee7b10').add_to(folium_map)

folium_map.save('./output/map4.html')
folium_map
```
